# Matplotlib Practice

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,10,20)
print(x)
```

```python
y = np.random.randint(1,50,20)
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color= 'b', marker = 'o', linestyle = '--')
```

```python
plt.plot(x,y, color= 'b', marker = 'o', linestyle = '--', markerfacecolor = 'r')
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,y, color= 'b', marker = 'o', linestyle = '--', markerfacecolor = 'r')
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,y, color= 'b', marker = 'o', linestyle = '--', markerfacecolor = 'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Demo Plot')
plt.show()
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, linewidths=5)
```
