## Subplots

```python
x
```

```python
y
```

```python
y2 = x*y
```

```python
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(x,y, marker = '*', color = 'r')

plt.subplot(1,2,2)
plt.plot(x,y2, marker = 'o', markersize = 10)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.6, 0.4, 0.3])

ax1.plot(x, y, 'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Y Plot')

ax2.plot(x, y2, 'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Y2 Plot')
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,6))

ax[0].plot(x,y,x,y2)
ax[0].set_xlim(0,8)

ax[1].plot(x,y**2, 'k')
ax[1].set_xticks(range(11))
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,6))

ax[0].plot(x,y,x,y2)
ax[0].set_xlim(0,8)

ax[1].plot(x,y**2, 'k')
ax[1].set_xticks([2,4,6,8])
ax[1].set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$'])
plt.show()
```

## Pie Chart

```python
labels = ['Cats', 'Dogs', 'Cows', 'Lions', 'Parrots']
sizes = [20, 40, 100, 150, 10]
```

```python
fig, ax = plt.subplots( figsize = (5, 5), dpi = 100)
ax.pie(sizes, labels = labels)
plt.show()
```

```python
fig, ax = plt.subplots( figsize = (5, 5), dpi = 100)
ax.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.show()
```

```python
fig, ax = plt.subplots( figsize = (5, 5), dpi = 100)
explode = (0.4, 0,0,0, 0)
ax.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = True, explode = explode)
plt.show()
```

```python
fig, ax = plt.subplots( figsize = (5, 5), dpi = 100)
explode = (0.4, 0,0,0, 0)
ax.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = True, explode = explode)
plt.show()
```
