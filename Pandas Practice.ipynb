```python
# reading data using .items()

# first variable has the column names
# second variable has the respective row values 

for first, second in pokemon_data.head(2).items():
    print("first====>>>>", first)
    print("second====>>>>", second)
    print("=================================")
    
```

```python
# reading data using .iterrows()

# first variable has the row indexes
# second variable has the respective row values (as column name as key and value as respective value)

for first, second in pokemon_data.head(2).iterrows():
    print("first====>>>>", first)
    print("second====>>>>", second)
    print("=================================")
    
```

```python
# reading data using .itertuples()

# row variable has the rows as tuples

for row in pokemon_data.head(2).itertuples():
    print("rows here====>>>>", row)
    #print("second====>>>>", second)
    print("=================================")
    
```
